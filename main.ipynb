*****Cleaning Data*****

1. URLs (http)
2. mentions (@)
3. hashtags (#)
4. special letters
5. punctuations

```python
df['Resume'][0]
```

```python
def CleanResume(txt):
    cleanTXT = re.sub('http\S+\s', ' ', txt)  # removing http and all attached letters (\s for \)
    cleanTXT = re.sub('RT|cc', ' ', cleanTXT) # removes common Twitter-related terms often found in retweets or mentions.
    cleanTXT = re.sub('@\S+', '  ', cleanTXT) # removing @
    cleanTXT = re.sub('#\S+\s', ' ', cleanTXT) # removing #
    cleanTXT = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', cleanTXT) # removes special characters
    cleanTXT = re.sub(r'[^\x00-\x7f]', ' ', cleanTXT) #  removes non-ASCII characters (that are not in range [^\x00-\x7f])
    cleanTXT = re.sub('\s+', ' ', cleanTXT) # removes extra whitespace from the text (\r \n \t)
    return cleanTXT
```

```python
CleanResume("my ###grgrgr # % $  ^  ##### website is http://ABCDEFG.com and my email is @gmail hai")
```

```python
df['Resume'] = df['Resume'].apply(lambda x: CleanResume(x))
```

```python
df['Resume'][0]
```

```python
le = LabelEncoder()
```

```python
df['Category'] = le.fit_transform(df['Category'])
```

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
requiredTXT = tfidf.fit_transform(df['Resume'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(requiredTXT, df['Category'], test_size=0.2, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)

# 7351 is the sparse array created by tfidf
```

```python
clf = OneVsRestClassifier(KNeighborsClassifier())
clf.fit(X_train, y_train)
ypred = clf.predict(X_test)
print(ypred)
print(accuracy_score(y_test, ypred))
```

```python
pickle.dump(tfidf, open('tfidf.pkl', 'wb'))
pickle.dump(clf, open('clf.pkl', 'wb'))
```
